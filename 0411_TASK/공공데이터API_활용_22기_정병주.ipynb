```python
datas = []

for i in range(len(te)):
    deal_fee = te[i].거래금액.string.strip()
    deal_type = te[i].거래유형.string.strip()
    built_yr = te[i].건축년도.string.strip()
    dong_name = te[i].법정동.string.strip()
    apt_name = te[i].아파트.string.strip()
    month = te[i].월.string.strip()
    day = te[i].일.string.strip()
    size = te[i].전용면적.string.strip()
    place = te[i].중개사소재지.string.strip()
    gu_code = te[i].지역코드.string.strip()
    
    data = [deal_fee, deal_type, built_yr, dong_name, apt_name, month, day, size, place, gu_code]
    datas.append(data)
```

```python
import pandas as pd
df = pd.DataFrame(datas, columns=['deal_fee', 'deal_type', 'built_yr', 'dong_name', 'apt_name', 'month', 'day', 'size', 'place', 'gu_code'])
```
