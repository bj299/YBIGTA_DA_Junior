### Subword Tokenizer - 자연어처리 모델을 훈련할 때 tokenizing 된 단어의 개수(단어사전, vocab)는 모델 성능에 다양한 영향을 미치게 된다. 기계가 모르는 단어로 인해 문제를 푸는 것이 까다로워지는 상황(Out-of-Vocabulary)이 나타나는데 이를 해결하기 위한 방안이 Subword Segmentation이다. Subword Segmentation은 하나의 단어는 더 작은 단위의 의미있는 여러 subword(책+가방)의 조합으로 구성되는 경우가 많으므로, 하나의 단어를 여러 subword로 분리해서 단어를 인코딩/임베딩 하겠다는 의도의 전처리 작업이며 여기서도 단어 집합의 크기를 잘 설정하는 것이 필요하다. 
- BPE: Byte Pair Encoding으로 기본적으로 연속적으로 가장 많이 등장한 글자의 쌍을 찾아서 하나의 글자로 병합하는 방식을 수행한다. 기존에는 단어의 빈도수로 이루어진 단어 집합을 활용하였다면 BPE 알고리즘을 적용할 경우 모든 단어들을 글자 단위로 분리하고 가장 빈도수가 높은 유니그램의 쌍을 하나의 유니그램으로 통합하는 과정을 반복한다. 즉, 단어보다 더 작은 단위로 인코딩을 가능하게 한다.
- SentencePiece: 빈도수를 기반으로 BPE를 수행하는 방식으로 기타 Subword Tokenizing Algorithm들도 내장하고 있다. 사전 토큰화 작업없이 단어 분리 토큰화를 수행하므로 특정 언어에 종속되지 않는 장점이 있다.
- WPM: Word Piece Model로 BPE의 변형 알고리즘이다. 병합되었을 때 Corpus의 우도(Likelihood)를 가장 높이는 쌍을 기준으로 병함이 이루어진다. 최신 딥러닝 모델 중 하나인 BERT를 훈련하기 위해서 사용하기도 한다.
